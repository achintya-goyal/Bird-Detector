```python
import os
from glob import glob
from tokenize import group

import pandas as pd
from functools import reduce
from xml.etree import ElementTree as et
```

```python
xml_list = glob('./data_images/*.xml')

xml_list = list(map(lambda x: x.replace('\\','/'), xml_list))
```

```python
xml_list
```

```python
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()

    image_name = root.find('filename').text
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text
    objs = root.findall('object')
    parser = []

    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        xmax = int(bndbox.find('xmax').text)
        ymin = int(bndbox.find('ymin').text)
        ymax = int(bndbox.find('ymax').text)
        parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])

    return parser
```

```python
parser_all = list(map(extract_text, xml_list))
```

```python
parser_all
```

```python
data = reduce(lambda x, y: x + y, parser_all)
```

```python
data
```

```python
df = pd.DataFrame(data, columns=['filename', 'width', 'height', 'name', 'xmin', 'xmax', 'ymin', 'ymax'])
```

```python
df.head()
```

```python
cols = ['width', 'height', 'xmin', 'xmax', 'ymin', 'ymax']
df[cols] = df[cols].astype(int)
df.info()
```

```python
df['center_x'] = ((df['xmin'] + df['xmax']) / 2)/df['width']
df['center_y'] = ((df['ymin'] + df['ymax']) / 2)/df['height']
df['w'] = (df['xmax'] - df['xmin']) / df['width']
df['h'] = (df['ymax'] - df['ymin']) / df['height']
```

```python
df.head()
```

```python
images = df['filename'].unique()
len(images)
```

```python
img_df = pd.DataFrame(images, columns=['filename'])
img_train = tuple(img_df.sample(frac=0.85)['filename'])
```

```python
img_train
```

```python
img_test = tuple(img_df.query(f'filename not in {img_train}')['filename'])
img_test
```

```python
len(img_train), len(img_test)
```

```python
train_df = df.query(f'filename in {img_train}')
test_df  = df.query(f'filename in {img_test}')
```

```python
def label_encode(x):
    labels = {'pigeon':0,'hen':1,'eagle':2,'owl':3}
    return  labels[x]
```

```python
train_df.loc[:, 'id'] = train_df['name'].apply(label_encode)
test_df.loc[:, 'id'] = test_df['name'].apply(label_encode)
```

```python
train_df.head()
```

```python
from shutil import move
```

```python
train_folder = 'data_images/train'
test_folder = 'data_images/test'

os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python
cols = ['filename', 'id', 'center_x', 'center_y', 'w', 'h']
group_by_obj_train = train_df[cols].groupby('filename')
group_by_obj_test = test_df[cols].groupby('filename')
```

```python
def save_data(filename, folder_path, group_obj):
    src = os.path.join('data_images', filename)
    dst = os.path.join(folder_path, filename)
    move(src, dst)

    text_filename = os.path.join(folder_path, os.path.splitext(filename)[0] + '.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename, header=False, index=False, sep = ' ')
```

```python
group_by_obj_train.groups.keys()
```

```python
filename_series = pd.Series(group_by_obj_train.groups.keys())
filename_series
```

```python
filename_series.apply(save_data, args=(train_folder, group_by_obj_train))
```

```python
filename_series_test = pd.Series(group_by_obj_test.groups.keys())
```

```python
filename_series_test.apply(save_data, args=(test_folder, group_by_obj_test))
```
